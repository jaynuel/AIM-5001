```python
#Answer to question 4. 
#To print the titles in bold letters I added: "\033[1m"+ before the text and after the text, I added "\033[0m").
#To leave a space between the first title and the answer I added a space by printing an empty row('').

print("\033[1m" +'Answer to question 4:'+"\033[0m")
print('')
print("\033[1m" +'Below is the list of the commulative miles traveled per day:'+"\033[0m")



# Total miles from NYC to Key West FL = 1,436
# Information optained from google maps; here is the link: https://www.google.com/maps/dir/New+York+City,+NY/Key+West,+Florida+33040/
#@32.3435157,-86.9035371,5z/data=!3m1!4b1!4m14!4m13!1m5!1m1!1s0x89c24fa5d33f083b:0xc80b8f06e177fe62!2m2!1d-74.
#0059728!2d40.7127753!1m5!1m1!1s0x88d1b134ad952377:0x3fcee92f77463b5e!2m2!1d-81.7799871!2d24.5550593!3e0

#The array travel_distance contains the amount of miles traveled per day, totaling the 1,436 miles from NYC 
#to Key West FL.

#To add the commulative miles, I used the cumsum numpy function and assigned the variable cummulative to the result. 

travel_distance= np.array([55,65,0,28,45,17,50,39,26,0,26,43,12,63,60,34,67,0,21,12,
                           0,36,51,45,47,65,60,66,68,59,55,67,65,70,19])
comulative= np.cumsum(travel_distance)

#I decided that I wanted to use a dictionary object to show the traveled miles per day. 
#I started by creating and empty dictionary called distance_per_day.
#I created a list containing the keys I will used in my dictionary.

distance_per_day={}
days=['Day_1','Day_2','Day_3','Day_4','Day_5','Day_6','Day_7','Day_8','Day_9','Day_10','Day_11',
      'Day_12','Day_13','Day_14','Day_15','Day_16','Day_17','Day_18','Day_19','Day_20','Day_21',
      'Day_22','Day_23','Day_24','Day_25','Day_26','Day_27','Day_28','Day_29','Day_30','Day_31',
      'Day_32','Day_33','Day_34','Day_35']


#I created a for loop to iterate the "comulative" array and the "days" list, which contains the new keys that will be
#used in the dictionary. I used the .pop to replace the old keys for the new keys.  

for i in range(35):
    distance_per_day[days[i]]=distance_per_day.pop(0, comulative[i])

#I formated the way in which the dictionary will be printed using a for loop with the function ".items".   
    
    
for key, value in distance_per_day.items():
    print(key, ':', value)


#The results show the commulative miles traveled per day.    










```
